```python
# Finding the distributions and Checking for Outliers of data in categorical column Versus Numerical Columns
## Numerical_Columns
for col in numerical_cols: 
    plt.figure(figsize=(10,8))
    sns.histplot(job_df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
## Check if there any outliers in Numerical Columns:
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=job_df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
## Check for distributions in Categorical Columns
for col in categorical_cols:
    top_n = job_df[col].value_counts().nlargest(10)
    
    plt.figure(figsize=(10,8))
    sns.barplot(x=top_n.index, y=top_n.values)
    plt.title(f"Top categories in {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# PART C: Answer Key Business Questions
## Q1. What is the average salary 
Average_salary = job_df['salary'].mean().round(0)
Average_salary
```

```python
## Q2. How does the average salary compare to the level of education
salary_education_comparison = job_df.groupby('education_level')['salary'].mean().sort_values(ascending = False).round()
salary_education_comparison

## Result; Employees with the Highest Eduction level(PhD) earned the most compared to those with lower levels of education
```

```python
## Q3. How does the average salary compare to the Level of experience
salary_level_comparison = job_df.groupby('experience_years')['salary'].mean().sort_values(ascending = False)
salary_level_comparison

## Result: Employees with the highest levels of Expereince(over 10years) earned the most amount of money compared to those with lower experience in years,
```

```python
## Is there any correlation between experience in years and the salary?
exper_salary_corr = job_df['experience_years'].corr(job_df['salary'])
print(exper_salary_corr)
```

```python
## Q4. WHICH INDUSTRY AND JOB TITLE earn the highest salary
most_salary = (job_df.groupby(['job_title', 'industry'])['salary'].mean().sort_values(ascending = False).reset_index())
most_salary

## Result: AI Engineers working in Media, Healthcare and Finance are the Highest earners with a salary of 174K
```

```python
## To view the bottom 10 Job Titles and Industry with the lowest salary
most_salary.tail(10)

## Data Analysts working in Consulting, Healthcare and Manufacturing are the least earners.
```

```python
## Q5.How does having a certification affect salary whean controlling for experience in years?
certs_salary = (job_df.groupby(['experience_years', 'certifications'])['salary'].mean().reset_index())
plt.figure(figsize=(10,6))
sns.lineplot(
    data= certs_salary,
    x='experience_years',
    y='salary',
    hue='certifications'
)

plt.title('Average Salary Growth by Experience Controlling for Certifications')
plt.show()


## Result: Employees who had numerous certifications earned a hihgher salary compared to those with Zero on One certifications.
```

```python
## Q7.Find the breakdown of companies_size. Which sub category in the size of company is the most dominant?
company_breakdown = job_df.groupby('company_size')['company_size'].value_counts()
company_breakdown

## Result: Employees are equally distributed across all the 4 categories in regards to the company size. No single category is SIGNIFICANTLY DOMINANT over the rest.
```

```python
## Q8. How are the various Job titles and Industries spread across the locations
## Create a frequency table to view our data for Job Title Versus Location
job_loc = pd.crosstab(job_df['location'], job_df['job_title'])
print(job_loc)
```

```python
## Q8. How are the job titles spread across the locations
plt.figure(figsize=(12,8))
sns.heatmap(job_loc, cmap='coolwarm')

plt.title('Job Title Distribution Across Locations')
plt.xlabel('job_title')
plt.ylabel('location')
plt.show()
```

```python
## For indusry versus Location
indu_loc = pd.crosstab(job_df['location'], job_df['industry'])
print(indu_loc)
```

```python
## To visualize our Industry vs location using a heat map
plt.figure(figsize=(12,8))
sns.heatmap(indu_loc, cmap='viridis')

plt.title('Industry Distribution Across Locations')
plt.xlabel('industry')
plt.ylabel('location')
plt.show()
```
